1. "What is hypothesis testing in statistics? - Hypothesis testing is a method to make inferences about population parameters using sample data."

2. "What is the null hypothesis, and how does it differ from the alternative hypothesis? - Null hypothesis (H0) assumes no effect; alternative hypothesis (H1) assumes an effect or difference."

3. "What is the significance level in hypothesis testing, and why is it important? - Significance level (α) is the threshold for rejecting H0, commonly set at 0.05."

4. "What does a P-value represent in hypothesis testing? - P-value is the probability of observing the data assuming H0 is true."

5. "How do you interpret the P-value in hypothesis testing? - If P-value ≤ α, reject H0; otherwise, fail to reject H0."

6. "What are Type 1 and Type 2 errors in hypothesis testing? - Type I error: rejecting true H0; Type II error: failing to reject false H0."

7. "What is the difference between a one-tailed and a two-tailed test in hypothesis testing? - One-tailed tests look for effect in one direction, two-tailed in both directions."

8. "What is the Z-test, and when is it used in hypothesis testing? - Z-test is used when population variance is known and sample size is large (n > 30)."

9. "How do you calculate the Z-score, and what does it represent in hypothesis testing? - Z = (X̄ - μ) / (σ/√n); it shows how many standard deviations the sample mean is from μ."

10. "What is the T-distribution, and when should it be used instead of the normal distribution? - T-distribution is used when sample size is small (n < 30) and population variance is unknown."

11. "What is the difference between a Z-test and a T-test? - Z-test uses known population variance; T-test uses sample variance."

12. "What is the T-test, and how is it used in hypothesis testing? - T-test compares sample mean(s) to test hypotheses about population mean(s)."

13. "What is the relationship between Z-test and T-test in hypothesis testing? - Both test means; Z-test is for large samples or known σ, T-test is for small samples or unknown σ."

14. "What is a confidence interval, and how is it used to interpret statistical results? - Confidence interval gives a range of plausible values for a parameter, with a specified probability."

15. "What is the margin of error, and how does it affect the confidence interval? - Margin of error defines the width of the confidence interval and reflects sampling variability."

16. "How is Bayes' Theorem used in statistics, and what is its significance? - Bayes' Theorem updates the probability of a hypothesis based on new evidence."

17. "What is the Chi-square distribution, and when is it used? - Chi-square distribution is used for categorical data and tests of independence or goodness-of-fit."

18. "What is the Chi-square goodness of fit test, and how is it applied? - It tests whether observed categorical data match an expected distribution."

19. "What is the F-distribution, and when is it used in hypothesis testing? - F-distribution is used to compare variances and in ANOVA tests."

20. "What is an ANOVA test, and what are its assumptions? - ANOVA tests whether means of 3+ groups are equal, assuming normality and equal variances."

21. "What are the different types of ANOVA tests? - One-way ANOVA (1 factor), Two-way ANOVA (2 factors), and repeated measures ANOVA."

22. "What is the F-test, and how does it relate to hypothesis testing? - F-test compares variances or multiple group means in hypothesis testing."





```python
# Cell for all 27 statistical tasks in Python - Google Colab compatible
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Z-test for sample mean vs known population mean
sample = np.random.normal(loc=52, scale=10, size=30)
pop_mean = 50
z = (np.mean(sample) - pop_mean) / (np.std(sample, ddof=1) / np.sqrt(len(sample)))
p = 2 * (1 - stats.norm.cdf(abs(z)))
print(f"1. Z-test: z = {z:.3f}, p = {p:.3f}")

# 2. Simulate data and calculate P-value
data = np.random.normal(100, 15, 100)
z = (np.mean(data) - 100) / (np.std(data, ddof=1) / np.sqrt(len(data)))
p = 2 * (1 - stats.norm.cdf(abs(z)))
print(f"2. Simulated Z-test: z = {z:.3f}, p = {p:.3f}")

# 3. One-sample Z-test
def one_sample_z_test(data, mu):
    z = (np.mean(data) - mu) / (np.std(data, ddof=1) / np.sqrt(len(data)))
    p = 2 * (1 - stats.norm.cdf(abs(z)))
    return z, p
z, p = one_sample_z_test(sample, 50)
print(f"3. One-sample Z-test: z = {z:.3f}, p = {p:.3f}")

# 4. Two-tailed Z-test visualization
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)
plt.plot(x, y)
plt.fill_between(x, y, where=(x <= -abs(z)) | (x >= abs(z)), color='red', alpha=0.3)
plt.axvline(x=z, color='blue', linestyle='--')
plt.title("4. Two-tailed Z-test")
plt.grid(); plt.show()

# 5. Type I and II errors visualization
def visualize_errors():
    x = np.linspace(-4, 4, 1000)
    h0 = stats.norm.pdf(x, 0, 1)
    h1 = stats.norm.pdf(x, 1.5, 1)
    plt.plot(x, h0, label='H0')
    plt.plot(x, h1, label='H1')
    plt.fill_between(x, h0, where=(x > 1.96), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, h1, where=(x <= 1.96), color='orange', alpha=0.3, label='Type II Error')
    plt.axvline(1.96, color='black', linestyle='--')
    plt.legend(); plt.title("5. Type I and II Errors"); plt.grid(); plt.show()
visualize_errors()

# 6. Independent T-test
a = np.random.normal(5, 2, 30)
b = np.random.normal(6, 2, 30)
t, p = stats.ttest_ind(a, b)
print(f"6. Independent T-test: t = {t:.3f}, p = {p:.3f}")

# 7. Paired T-test
before = np.random.normal(100, 10, 30)
after = before + np.random.normal(5, 5, 30)
t, p = stats.ttest_rel(before, after)
print(f"7. Paired T-test: t = {t:.3f}, p = {p:.3f}")

# 8. Compare Z-test vs T-test
z, p_z = one_sample_z_test(a, 5)
t, p_t = stats.ttest_1samp(a, 5)
print(f"8. Z-test vs T-test: Z-p = {p_z:.3f}, T-p = {p_t:.3f}")

# 9. Confidence Interval
def confidence_interval(data, conf=0.95):
    mean = np.mean(data)
    se = stats.sem(data)
    h = se * stats.t.ppf((1 + conf) / 2., len(data)-1)
    return mean - h, mean + h
ci = confidence_interval(a)
print(f"9. Confidence Interval: {ci}")

# 10. Margin of Error
moe = (ci[1] - ci[0]) / 2
print(f"10. Margin of Error: {moe:.3f}")

# 11. Bayes' Theorem
def bayes_theorem(prior_A, prob_B_given_A, prob_B):
    return (prob_B_given_A * prior_A) / prob_B
posterior = bayes_theorem(0.01, 0.95, 0.05)
print(f"11. Bayesian Posterior: {posterior:.4f}")

# 12. Chi-square test of independence
from scipy.stats import chi2_contingency
obs = np.array([[10, 20], [20, 40]])
chi2, p, _, _ = chi2_contingency(obs)
print(f"12. Chi-square Independence: χ2 = {chi2:.3f}, p = {p:.3f}")

# 13. Expected frequencies
_, _, expected, _ = chi2_contingency(obs)
print(f"13. Expected Frequencies:\n{expected}")

# 14. Goodness of fit test
obs = [20, 30, 50]
exp = [25, 25, 50]
chi2, p = stats.chisquare(obs, exp)
print(f"14. Goodness of Fit: χ2 = {chi2:.3f}, p = {p:.3f}")

# 15. Simulate Chi-square distribution
x = np.linspace(0, 20, 1000)
plt.plot(x, stats.chi2.pdf(x, df=2))
plt.title("15. Chi-square Distribution"); plt.grid(); plt.show()

# 16. F-test (variances)
f = np.var(a, ddof=1) / np.var(b, ddof=1)
df1, df2 = len(a)-1, len(b)-1
p = 1 - stats.f.cdf(f, df1, df2)
print(f"16. F-test: F = {f:.3f}, p = {p:.3f}")

# 17. ANOVA test (one-way)
group1 = np.random.normal(5, 1, 20)
group2 = np.random.normal(6, 1, 20)
group3 = np.random.normal(7, 1, 20)
f, p = stats.f_oneway(group1, group2, group3)
print(f"17. One-way ANOVA: F = {f:.3f}, p = {p:.3f}")

# 18. One-way ANOVA + plot
plt.boxplot([group1, group2, group3])
plt.title("18. One-way ANOVA Boxplot")
plt.xticks([1, 2, 3], ['G1', 'G2', 'G3']); plt.grid(); plt.show()

# 19. Check ANOVA assumptions (Shapiro for normality)
for i, g in enumerate([group1, group2, group3], start=1):
    stat, p = stats.shapiro(g)
    print(f"19. Group {i} normality p = {p:.3f}")

# 20. Two-way ANOVA (using statsmodels)
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
df = pd.DataFrame({
    'score': np.random.randn(60),
    'factor1': np.repeat(['A', 'B'], 30),
    'factor2': np.tile(np.repeat(['X', 'Y'], 15), 2)
})
model = ols('score ~ factor1 * factor2', data=df).fit()
anova = sm.stats.anova_lm(model, typ=2)
print("20. Two-way ANOVA:\n", anova)

# 21. F-distribution visualization
x = np.linspace(0, 5, 1000)
plt.plot(x, stats.f.pdf(x, 5, 10))
plt.title("21. F-distribution"); plt.grid(); plt.show()

# 22. One-way ANOVA boxplots
plt.boxplot([group1, group2, group3])
plt.title("22. One-way ANOVA Boxplots"); plt.grid(); plt.show()

# 23. Simulate normal data + test mean
data = np.random.normal(100, 10, 50)
t, p = stats.ttest_1samp(data, 100)
print(f"23. Hypothesis Test on Simulated Mean: t = {t:.3f}, p = {p:.3f}")

# 24. Chi-square test for variance
sample_var = np.var(data, ddof=1)
chi2 = (len(data) - 1) * sample_var / 100
p = 1 - stats.chi2.cdf(chi2, df=len(data)-1)
print(f"24. Chi-square Test for Variance: χ2 = {chi2:.3f}, p = {p:.3f}")

# 25. Z-test for proportions
p1, n1 = 0.55, 100
p2, n2 = 0.5, 100
pooled = (p1*n1 + p2*n2) / (n1 + n2)
z = (p1 - p2) / np.sqrt(pooled * (1 - pooled) * (1/n1 + 1/n2))
p = 2 * (1 - stats.norm.cdf(abs(z)))
print(f"25. Z-test for Proportions: z = {z:.3f}, p = {p:.3f}")

# 26. F-test + visualization
f = np.var(group1, ddof=1) / np.var(group2, ddof=1)
df1, df2 = len(group1)-1, len(group2)-1
p = 1 - stats.f.cdf(f, df1, df2)
x = np.linspace(0, 5, 1000)
plt.plot(x, stats.f.pdf(x, df1, df2))
plt.axvline(f, color='red', linestyle='--')
plt.title("26. F-test Visualization"); plt.grid(); plt.show()
print(f"26. F = {f:.3f}, p = {p:.3f}")

# 27. Chi-square goodness of fit with simulation
obs = np.random.randint(10, 50, 5)
exp = np.full(5, np.mean(obs))
chi2, p = stats.chisquare(obs, exp)
print(f"27. Simulated Goodness of Fit: χ2 = {chi2:.3f}, p = {p:.3f}")

```
