  17.Describe the time complexity of accessing elements in a dictionary?
     ->In Python, dictionaries are implemented using hash tables, which allow
       for efficient access to elements.
        Average Case: O(1):=In most cases, accessing an element by its key in a dictionary is constant time, i.e., O(1). This means the time it takes to access a value does not depend on the size of the dictionary
        
        Worst Case: O(n):=In rare cases, when there are many collisions (two or more keys having the same hash), the dictionary may fall back to linear search to resolve these collisions, resulting in O(n) time complexity.

  20.Explain the difference between a list and a dictionary in terms of data retrieval.
     ->List: Data is retrieved by index (position), meaning you access elements in a specific order.

      Dictionary: Data is retrieved by key, allowing fast lookups but with no inherent order.

      Performance: Lists require sequential searching (O(n)) for non-indexed access, while dictionaries offer constant time lookup (O(1)) for key-based access.
