1.What is NumPy, and why is it widely used in Python?
->NumPy is a Python library for efficient numerical and scientific computing, centered around its powerful ndarray object. It supports fast, vectorized operations and advanced math functions like linear algebra and statistics. Widely used in data science and machine learning, it serves as the foundation for many other libraries like Pandas and TensorFlow.


2. How does broadcasting work in NumPy?
->Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes by automatically expanding the smaller array to match the shape of the larger one without copying data.

3.A What is a Pandas DataFrame?

->A Pandas DataFrame is a 2-dimensional labeled data structure in Python, similar to a table in a database or an Excel spreadsheet.
It consists of rows and columns, where each column can hold different data types (e.g., int, float, string).
DataFrames are widely used for data analysis, manipulation, and cleaning in Python.

4. Explain the use of the groupby() method in Pandas?
->The groupby() method in Pandas is used to split a DataFrame into groups based on one or more columns, allowing you to perform aggregate operations (like sum, mean, count) or apply custom functions on each group separately.

5. Why is Seaborn preferred for statistical visualizations?

6.What are the differences between NumPy arrays and Python lists?
->NumPy arrays store elements of the same data type and support fast, vectorized numerical operations. Python lists can hold mixed data types but are slower and less memory-efficient. NumPy is ideal for numerical computing, while lists are more flexible for general use.

7.What is a heatmap, and when should it be used?

A heatmap is a graphical representation of data where values are shown as colors in a matrix or grid. It’s used to visualize patterns, correlations, or intensity in large datasets, making it easy to spot trends or outliers.

8. What does the term “vectorized operation” mean in NumPy?
->A vectorized operation in NumPy means performing element-wise operations on entire arrays without explicit Python loops, using optimized, low-level code. This allows for faster and more efficient computations by applying operations simultaneously across all elements.

9. How does Matplotlib differ from Plotly?

Plotly is an interactive, web-based plotting library that supports 2D and 3D plots with zooming, hovering, and dynamic updates.

17. Explain the difference between apply() and map() in Pandas?
->apply() can be used on DataFrames or Series to apply a function across rows, columns, or elements, allowing complex operations on entire rows or columns.

map() is used only on Series to map each value to another value (e.g., via a dictionary or function), mostly for element-wise transformations.

 21.Why is NumPy’s array slicing faster than Python’s list slicing?

2.Write a Pandas script to find the mean of a specific column in a DataFrame?


```python
import pandas as pd
data={
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df=pd.DataFrame(data)

print(df['Age'].mean())
```

4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Math': [90, 85, 70, 65, 95],
    'Science': [88, 82, 75, 60, 98],
    'English': [78, 80, 85, 70, 75]
})

c_matrix=data.corr()

sns.heatmap(c_matrix,annot=True)
plt.title("Heat Map")
plt.show()
```

6. Create a DataFrame and add a new column based on an existing column



```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Score': [85, 92, 78]
}

df = pd.DataFrame(data)

df['Grade'] = df['Score'].apply(lambda x: 'A' if x >= 90 else 'B' if x >= 80 else 'C')

print(df)
```

8.Create a line plot with multiple lines using Matplotlib?


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y1, label='Line 1', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', color='green', marker='s')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')
plt.legend()

plt.show()
```

9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold?

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Marks': [85, 67, 92, 74]
}
df = pd.DataFrame(data)

filtered_df = df[df['Marks'] > 75]

print(filtered_df)
```

10.Create a histogram using Seaborn to visualize a distribution?

```python
import seaborn as sns
import matplotlib.pyplot as plt

scores = [56, 65, 70, 72, 75, 78, 80, 82, 85, 88, 90, 92, 95, 98]

sns.histplot(scores, bins=6, kde=True, color='skyblue')

plt.title("Score Distribution")
plt.xlabel("Scores")
plt.ylabel("Frequency")

plt.show()
```

11.Perform matrix multiplication using NumPy?

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = np.dot(A, B)

print("Matrix Multiplication Result:")
print(result)
```

13. Create a 3D scatter plot using Plotly.


```python
import plotly.express as px
import pandas as pd

# Sample data
df = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [10, 11, 12, 13, 14],
    'z': [5, 4, 3, 2, 1],
    'color': ['A', 'B', 'A', 'B', 'A']
})

# Create 3D scatter plot
fig = px.scatter_3d(df, x='x', y='y', z='z', color='color', title='3D Scatter Plot')
fig.show()
```
